# LAUS  
This outlines how to retrieve labor force statistics from the Local Area Unemployment Statistics via JobsEQ.  

Beginning at data explorer, select annual data. Select "Region" and populate with US, Tennessee, GNRC, MPO, GNRC Region, and then under GNRC Region "Counties". Select "Year/Month" data and choose your range (2000 - recent) and then select the value "Unemployment Rate".  

When you've downloaded this csv, open it and replace the title row with the year (as all of the values are unemployment rates), delete the header and data notes, rename region to "NAME", and select all of the rates and make them "General" format. You can also quickly delimit the NAMEs to remove the ", Tennessee" off of the end of the counties and change "USA" to "US". This can also be done here, of course.

```python
#LAUS
data = pd.read_csv('../../Data Downloads/JobsEQLAUS.csv')
```

```python
cols = data.columns
```

```python
data = data.melt(var_name = 'Year', id_vars = ['NAME'], value_vars = cols, value_name = 'Unemployment Rate')
```

```python
data['Unemployment Rate'] = data['Unemployment Rate'] * 100
```

From data explorer, select "education", and then select the GNRC region and all counties in it, all award levels, and only awards as the value. Format the document so that there is only one column and the commas are removed from the numbers, then save in Data Downloads as "JobsEQAwards_RegionCounty.csv".

```python
data = pd.read_csv('../../Data Downloads/JobsEQAwards_RegionCounty.csv')
```

```python
cols = data.columns
```

```python
data = data.melt(var_name = 'NAME', id_vars = ['Award Level', 'Year'], value_vars = cols, value_name = 'Awards')
```

```python
test = data['NAME'].str.split(r"\-|,", expand = True)
```

```python
data['NAME'] = test[0]
```

```python
data['NAME'] = data['NAME'].str.strip()
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
geos = ['GNRC Region', 'Cheatham County', 'Davidson County', 'Dickson County', 'Houston County', 'Humphreys County', 'Maury County', 'Montgomery County',
        'Robertson County', 'Rutherford County', 'Stewart County', 'Sumner County', 'Trousdale County', 'Williamson County', 'Wilson County'
       ]
```

```python
thing = data.set_index(['Award Level', 'Year', 'NAME']).transpose()
thing.head()
```

```python
thing = thing.fillna(0.0)
cols = thing.columns
thing[cols] = thing[cols].astype(float)
```

```python
for year in years:
    for geo in geos:
        thing[('Certificate < 2 Years', year, '{}'.format(geo))] = thing[('Certificate < 1 Yr', year, '{}'.format(geo))] + thing[('Certificate 1+ but < 2 Yr',
                                                                                                                                  year, '{}'.format(geo))]
        thing[('Certificate > 2 Years', year, '{}'.format(geo))] = thing[('Certificate 2+ but < 4 Yr', year, '{}'.format(geo))]
        thelist = [thing[("Master's", year, '{}'.format(geo))], thing[("Post-Master's", year, '{}'.format(geo))], 
                   thing[("Doctorate", year, '{}'.format(geo))]]
        thing[("Master's or Above", year, '{}'.format(geo))] = sum(thelist)
        thelist = [thing[("Bachelor's", year, '{}'.format(geo))], thing[("Post-Baccalaureate", year, '{}'.format(geo))]]
        thing[("Bachelor's", year, '{}'.format(geo))] = sum(thelist)
        thing = thing.drop(columns = [('Certificate < 1 Yr', year, '{}'.format(geo)), ('Certificate 1+ but < 2 Yr', year, '{}'.format(geo)), 
                                      ('Certificate 2+ but < 4 Yr', year, '{}'.format(geo)), ('Post-Baccalaureate', year, '{}'.format(geo))
                                     ])
```

```python
thing = thing.transpose().reset_index(drop = False)
```

```python
thing.to_csv('../../Outputs/JOBSEQ_AWARDS_REGIONCOUNTY.csv', index = False)
```
