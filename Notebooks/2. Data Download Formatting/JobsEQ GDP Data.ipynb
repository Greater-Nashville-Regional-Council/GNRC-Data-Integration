```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from functools import reduce
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 150)
def percentchange(x, y):
    try:
        return ((x - y)*100/y)
    except ZeroDivisionError:
        return 0
def realchange(x, y):
    return x-y
def percent(x, y):
    return (x/y)*100
```

From data explorer, select "GDP", and then select the GNRC region and all counties in it, then values: 'GDP', 'Taxes', 'Gross Surplus', 'Compensation'. Format the document so that there is only one column and the commas are removed from the numbers, then save in Data Downloads as "JobsEQGDP_RegionCounty.csv".

```python
data = pd.read_csv('../../Data Downloads/JobsEQGDP_RegionCounty.csv')
```

```python
data.head()
```

```python
data = data.set_index('NAME').transpose()
```

```python
data = data.rename(columns = {'NAME': 'Value'})
```

```python
data = data.reset_index(drop = False)
data.head()
```

```python
data['index'] = data['index'].str.strip()
data['Value'] = data['Value'].str.strip()
```

```python
boop = data['index'].str.split(pat = ".", expand = True)
data['Year'] = boop[0]
```

```python
data = data.drop(columns = 'index')
```

```python
data.head()
```

```python
data = data.rename(columns = {'USA': 'US', 'Cheatham County, Tennessee': 'Cheatham County', 'Davidson County, Tennessee': 'Davidson County', 
                              'Dickson County, Tennessee': 'Dickson County', 'Houston County, Tennessee': 'Houston County', 
                              'Humphreys County, Tennessee': 'Humphreys County', 'Maury County, Tennessee': 'Maury County', 
                              'Montgomery County, Tennessee': 'Montgomery County',
                              'Robertson County, Tennessee': 'Robertson County', 'Rutherford County, Tennessee': 'Rutherford County', 
                              'Stewart County, Tennessee': 'Stewart County', 'Sumner County, Tennessee': 'Sumner County', 
                              'Trousdale County, Tennessee': 'Trousdale County', 'Williamson County, Tennessee': 'Williamson County', 
                              'Wilson County, Tennessee': 'Wilson County'
                             })
```

```python
data.head()
```

```python
cols = data.columns
data = data.melt(var_name = 'NAME', id_vars = ['Value', 'Year'], value_vars = cols, value_name = 'Money')
```
